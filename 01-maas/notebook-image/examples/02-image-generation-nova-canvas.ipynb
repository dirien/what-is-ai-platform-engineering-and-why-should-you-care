# Image Generation with Amazon Nova Canvas

This notebook demonstrates how to generate images using Amazon Nova Canvas through the LiteLLM gateway.
Nova Canvas is a text-to-image model available via AWS Bedrock, exposed through the same OpenAI-compatible API
you use for chat completions.

## Prerequisites
- An API key from the MaaS portal with access to `amazon-nova-canvas`
- The `amazon-nova-canvas` model registered in LiteLLM

## 1. Setup and Configuration

```python
%pip install -q openai
```

```python
import os
import base64
from openai import OpenAI
from IPython.display import Image, display

# Default LiteLLM endpoint (in-cluster via the fixed-name litellm-lb service).
# Override with OPENAI_API_BASE env var if needed.
LITELLM_BASE_URL = os.environ.get('OPENAI_API_BASE', 'http://litellm-lb.maas.svc.cluster.local:4000/v1')
print(f"LiteLLM Endpoint: {LITELLM_BASE_URL}")

# Set your API key here (get it from the MaaS portal -> API Keys)
# os.environ['OPENAI_API_KEY'] = 'sk-your-api-key-here'

client = OpenAI(
    base_url=LITELLM_BASE_URL,
    api_key=os.environ.get('OPENAI_API_KEY', 'your-api-key')
)
```

## 2. Text-to-Image Generation

```python
# Generate an image from a text prompt
response = client.images.generate(
    model="amazon-nova-canvas",
    prompt="A futuristic city skyline at sunset, cyberpunk style, neon lights reflecting on water",
    size="1024x1024",
    n=1
)

# Decode and display the generated image
img_data = base64.b64decode(response.data[0].b64_json)
display(Image(data=img_data))
```

## 3. Multiple Prompts — Batch Generation

```python
prompts = [
    "A serene mountain landscape with a crystal-clear lake at dawn",
    "A portrait of a robot artist painting on an easel in a sunlit studio",
    "An abstract composition of geometric shapes in vibrant gradients"
]

for prompt in prompts:
    print(f"Generating: {prompt[:60]}...")
    response = client.images.generate(
        model="amazon-nova-canvas",
        prompt=prompt,
        size="1024x1024",
        n=1
    )
    img_data = base64.b64decode(response.data[0].b64_json)
    display(Image(data=img_data))
    print()
```

## 4. Saving Images to Disk

```python
# Generate and save an image to a file
response = client.images.generate(
    model="amazon-nova-canvas",
    prompt="A detailed technical diagram of a cloud architecture with Kubernetes pods",
    size="1024x1024",
    n=1
)

img_data = base64.b64decode(response.data[0].b64_json)

output_path = "generated_image.png"
with open(output_path, "wb") as f:
    f.write(img_data)

print(f"Image saved to: {os.path.abspath(output_path)}")
display(Image(filename=output_path))
```
